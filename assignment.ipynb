```python
import pandas as pd
import simplejson as json
```

```python
# reading the csv files

customer_one = pd.read_csv('customer_data_one.csv')
customer_two = pd.read_csv('customer_data_two.csv')

# finding the structure of the datasets

print(customer_one.shape)
print(customer_two.shape)
print(customer_one.columns)
print(customer_two.columns)
print(customer_one.dtypes)
print(customer_two.dtypes)
print(customer_one.head())
print(customer_two.head())
```

```python
# finding the number of duplicate rows and their values in the respective datasets

duplicates = customer_one.merge(customer_two, how='inner', on='mobile_number', suffixes=('_one', '_two'))
print(len(duplicates))

print(duplicates)
```

```python
# concatinating the two datasets and dropping all the duplicates except the last occurance and sorting the result by first_name
# a row is said to be duplicated if its corresponding mobile_number already exits

customer_data = pd.concat([customer_one,customer_two], ignore_index=True).drop_duplicates(subset='mobile_number', keep='last')

customer_data = customer_data.sort_values(by=['first_name'], ignore_index=True)

print(customer_data.head())
print(customer_data.columns)
print(customer_data.dtypes)

```

```python
# converting the final dataframe obtained after merge to json array of json objects and storing it in a file named customer_object.json

result_object = customer_data.to_json(orient='records')
result_object = json.loads(result_object)

output_file = open('customer_object.json', 'w')
json.dump(result_object, output_file, indent=4)
output_file.close()
```

```python
# converting the final dataframe obtained after merge to json array of json objects where primary key is specified to be the mobile_number in the schema and storing it in a file named customer_table.json

customer_data = customer_data.set_index('mobile_number')
result_table = customer_data.to_json(orient="table")
result_table = json.loads(result_table)

output_file = open('customer_table.json', 'w')
json.dump(result_table, output_file, indent=4)
output_file.close()

```
